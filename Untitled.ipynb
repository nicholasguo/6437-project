```python
import decode
import matplotlib
import matplotlib.pyplot as plt
```

```python
ciphertext = open("data/test/ciphertext.txt", "r").read()
```

```python
plaintext = open("data/test/plaintext.txt", "r").read()
```

```python
decoded, likelihoods, accepted, accuracies = decode.decode_plot(ciphertext, plaintext, False)
```

```python
plt = matplotlib.pyplot
plt.plot(likelihoods)
plt.axis([0, len(likelihoods), -50000, 0])
plt.ylabel('log likelihood')
plt.xlabel('iterations')
plt.title('Log Likelihood of accepted state')
plt.show()
```

```python
for i in range(1, len(accepted)):
    accepted[i] += accepted[i-1]
    
for i in range(len(accepted)-1, 50, -1):
    accepted[i] -= accepted[i-50]
    accepted[i] /= 50
    
for i in range(50):
    accepted[i] /= (i+1)
```

```python
plt = matplotlib.pyplot
plt.plot(accepted)
plt.axis([0, len(accepted), 0, 1])
plt.ylabel('acceptance rate')
plt.xlabel('iterations')
plt.show()
```

```python
plt = matplotlib.pyplot
plt.plot(accuracies)
plt.axis([0, len(likelihoods), 0, 1])
plt.ylabel('accuracy')
plt.xlabel('iterations')
plt.show()
```
