```python
import decode
import math
import matplotlib
import matplotlib.pyplot as plt
import time
```

```python
ciphertext = open("data/test/ciphertext_breakpoint.txt", "r").read()
```

```python
plaintext = open("data/test/plaintext.txt", "r").read()
```

```python
start = time.time()
guess = decode.decode(ciphertext, True)
end = time.time()
print(end - start)
decode.accuracy(guess, plaintext)
```

```python
permutation = "azfrtpvmlob.shwcqkuj exdyign"
decoded = decode.apply_perm(ciphertext, permutation)
prob = [decode.letter_probabilities[decode.DECODER[decoded[0]]]]
for i in range(1, len(ciphertext)):
    prob.append(decode.letter_transition_matrix[decode.DECODER[decoded[i]]][decode.DECODER[decoded[i-1]]])
    prob[-1] += prob[-2]
    
for i in range(len(ciphertext) - 1, 9, -1):
    prob[i] -= prob[i-10]
    
for i in range(10, len(ciphertext)):
    if prob[i] < -75:
        print(i)
        print(prob[i])
        break
```

```python
plt = matplotlib.pyplot
plt.plot(prob)
plt.axis([0, len(prob), -200, 0])
plt.ylabel('log likelihood')
plt.xlabel('iterations')
plt.title('Log Likelihood of accepted state')
plt.show()
```
